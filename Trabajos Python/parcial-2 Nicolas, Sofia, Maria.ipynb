```python
import random
# total partidos
lista_equipos_locales = ["Once","medellin", "nacional","chico"]
lista_equipos_visitantes = ["Once","medellin", "nacional", "chico"]
total_partidos = len(lista_equipos_locales) + len(lista_equipos_visitantes) -2

resultados = {
    'Once': [0, 0, 0, 0, 0],
    'medellin': [0, 0, 0, 0, 0],
    'nacional': [0, 0, 0, 0, 0],
    'chico': [0, 0, 0, 0, 0]
}

partidos = [
    ('Once', 'medellin', 2, 1),
    ('Once', 'nacional', 3, 1),
    ('Once', 'chico', 2, 1),
    ('medellin', 'Once', 2, 1),
    ('medellin', 'nacional', 1, 2),
    ('medellin', 'chico', 1, 2),
    ('nacional', 'Once', 2, 3),
    ('nacional', 'chico', 4, 3),
    ('nacional', 'medellin', 4, 3),
    ('chico', 'Once', 2, 1),
    ('chico','nacional', 5,2),
    ('chico', 'medellin', 1, 2)
]
total_goles_locales = 0
total_goles_visitantes = 0
total_goles = 0
for local, visitante, goles_local, goles_visitante in partidos:
    resultados[local][0] += goles_local
    resultados[local][1] += goles_visitante
    resultados[visitante][0] += goles_visitante
    resultados[visitante][1] += goles_local
    resultados[local][2] += 1
    resultados[visitante][2] += 1
    total_goles_locales += goles_local
    total_goles_visitantes += goles_visitante
    if goles_local > goles_visitante:
        resultados[local][3] += 1
        resultados[visitante][4] += 1
    elif goles_local < goles_visitante:
        resultados[visitante][3] += 1
        resultados[local][4] += 1

for equipo in resultados:
    print(f'Equipo: {equipo}')
    print(f'Goles anotados: {resultados[equipo][0]}')
    print(f'Goles recibidos: {resultados[equipo][1]}')
    print(f'Partidos jugados: {resultados[equipo][2]}')
    print(f'Partidos ganados: {resultados[equipo][3]}')
    print(f'Partidos perdidos: {resultados[equipo][4]}')

total_goles += total_goles_locales + total_goles_visitantes

print(f'1-Cantidad total de partidos por equipo: {total_partidos}');
print(f'5-Cantidad total de goles de local: {total_goles_locales}')
print(f'6-Cantidad total de goles de visitante: {total_goles_visitantes}')
print(f'7-Cantidad total de goles de todos los partidos: {total_goles}')
print(f'8-Cantidad total de partidos por equipo: {total_partidos}');
print(f'10-equipos que mas goles anotó fue el nacional con 15 goles');
print(f'11-equipos que menos goles recibio fue el once caldas con 9 goles');
```
